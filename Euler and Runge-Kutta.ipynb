# Euler's Method

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

$$\Bigg\{  \frac{\partial y}{\partial x} = 2xy \quad \textrm{and} \quad  y(0) = 1 $$ 


$$\int \frac{\partial y}{y} = \int 2xdx $$


$$ \ln|y| = x^2 + C \\ y = e^{x^2} $$


```python
def y(x):
    return np.exp(x**2)
def f(x, y):
    return 2*x*y
```

```python
a, b, N = 0, 1, 100
h = (b - a)/N
x = np.array([a + i*h for i in range(1, N)])
```

```python
plt.plot(x, y(x))
```

```python
u0 = 1
u = [u0]
for i in range(len(x)-1):
    u.append(f(x[i], u[i])*h + u[i])
```

```python
plt.plot(x, y(x), color='red')
plt.scatter(x, u)
```

```python
d0 = np.abs(y(x)-u)
d0
```

# Runge-Kutta's method

```python
def f(x, y):
    return 2*x*y
```

```python
z0 = 1
z = [z0]
for i in range(len(x)-1):
    z_ = (h/2)*f(x[i], z[i])+z[i]
    z.append(h*(f(x[i], z_)) + z[i])
```

```python
plt.plot(x, y(x), color='red')
plt.scatter(x, z)
```

```python
d1 = np.abs(y(x)-z)
d1
```

```python
np.mean(d0)
```

```python
np.mean(d1)
```
